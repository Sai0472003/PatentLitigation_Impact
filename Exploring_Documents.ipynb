```python
#Function to check if the first one or two words are uppercase
def starts_with_all_caps(text):
    if pd.isna(text):
        return False  # Handle NaN values safely
    words = text.split()  # Split the text into words
    if len(words) >= 2:
        return words[0].isupper() and words[1].isupper()  # Check if first two words are uppercase
    elif len(words) == 1:
        return words[0].isupper()  # Check if only one word exists and is uppercase
    return False

# Apply the filter to remove rows where the first one or two words are uppercase
filtered_docs = docs[~docs['long_description'].apply(starts_with_all_caps)]

filtered_docs
```

***Categortization of the Case Outcomes Based on NLP of document descriptions***


```python
#initializing case outcomes column
docs['case_outcome'] = "Uncategorized"

# 1. Plaintiff Wins
plaintiff_win_mask = docs['combined_description'].str.contains(
    r"plaintiff (prevails|wins|awarded|granted|successful|entitled to damages)",
    case=False, na=False, regex=True
)
docs.loc[plaintiff_win_mask, 'case_outcome'] = "Plaintiff Wins"

# 2. Defendant Wins
defendant_win_mask = docs['combined_description'].str.contains(
    r"defendant (prevails|wins|dismissed|granted|successful|not liable|case dismissed)",
    case=False, na=False, regex=True
)
docs.loc[defendant_win_mask, 'case_outcome'] = "Defendant Wins"

# 3. Case Dismissed
dismissed_mask = docs['combined_description'].str.contains(
    r"case (dismissed|closed) with( prejudice|out prejudice)?",
    case=False, na=False, regex=True
)
docs.loc[dismissed_mask, 'case_outcome'] = "Case Dismissed"

# 4. Settlement Reached
settlement_mask = docs['combined_description'].str.contains(
    r"settlement (reached|agreement|approved|finalized|filed)",
    case=False, na=False, regex=True
)
docs.loc[settlement_mask, 'case_outcome'] = "Settlement Reached"

# 5. Patent Declared Invalid
invalid_patent_mask = docs['combined_description'].str.contains(
    r"patent (invalidated|found invalid|void|unenforceable|revoked)",
    case=False, na=False, regex=True
)
docs.loc[invalid_patent_mask, 'case_outcome'] = "Patent Declared Invalid"

# 6. No Infringement Found
no_infringement_mask = docs['combined_description'].str.contains(
    r"no (infringement|violation) (found|determined|declared)",
    case=False, na=False, regex=True
)
docs.loc[no_infringement_mask, 'case_outcome'] = "No Infringement Found"

# 7. Pending Review
pending_review_mask = docs['combined_description'].str.contains(
    r"(PTAB|Patent Trial and Appeal Board|USPTO) (review|inter partes review|reexamination|pending)",
    case=False, na=False, regex=True
)
docs.loc[pending_review_mask, 'case_outcome'] = "Pending Review"

# 8. Mixed Outcome
mixed_outcome_mask = docs['combined_description'].str.contains(
    r"claims? (\d+[-–]\d+)? (infringed|valid) and claims? (\d+[-–]\d+)? (invalid|not infringed)",
    case=False, na=False, regex=True
)
docs.loc[mixed_outcome_mask, 'case_outcome'] = "Mixed Outcome"
```

```python
# Install necessary libraries
#!pip install transformers torch datasets

# Import required libraries
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load fine-tuned Legal-BERT model for classification
model_name = "zlucia/custom-legalbert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Move model to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Sample 5 categorized cases from the baseline for comparison
sampled_cases = docs.loc[docs['case_outcome'] != "Uncategorized", ['combined_description', 'case_outcome']].sample(5)

# Tokenize the selected text samples
inputs = tokenizer(sampled_cases['combined_description'].tolist(),
                   padding=True, truncation=True, max_length=512, return_tensors="pt")
inputs = {key: val.to(device) for key, val in inputs.items()}  # Move input tensors to GPU

# Run Legal-BERT for classification (Inference Mode)
with torch.no_grad():
    outputs = model(**inputs)
logits = outputs.logits

# Convert logits to probabilities
probs = torch.nn.functional.softmax(logits, dim=-1)

# Extract predicted labels (numerical)
predicted_labels = torch.argmax(probs, dim=-1).cpu().numpy()

# Retrieve Legal-BERT's label mapping
id2label = model.config.id2label  # Extracts predefined labels if available

# If the model lacks a proper label mapping, we define one manually
if not id2label or len(id2label) < 9:  # Ensure it maps to our 9 categories
    id2label = {
        0: "Plaintiff Wins",
        1: "Defendant Wins",
        2: "Case Dismissed",
        3: "Settlement Reached",
        4: "Patent Declared Invalid",
        5: "No Infringement Found",
        6: "Pending Review",
        7: "Mixed Outcome",
        8: "Other"
    }

# Convert numeric predictions to text labels
predicted_text_labels = [id2label[idx] for idx in predicted_labels]

# Attach predictions to the sampled dataframe
sampled_cases['LegalBERT_Prediction'] = predicted_text_labels

# Display the comparison using Pandas
from IPython.display import display
print("\n✅ Baseline vs. Legal-BERT Predictions\n")
display(sampled_cases)
```

```python
# Load Legal-BERT model for embeddings
legal_bert = SentenceTransformer("nlpaueb/legal-bert-base-uncased")

# Generate Legal-BERT embeddings
legal_embeddings = legal_bert.encode(docs_list, show_progress_bar=True)
```

```python
# Initialize and fit BERTopic with Legal-BERT embeddings
topic_model = BERTopic(embedding_model=legal_bert)
topics, probs = topic_model.fit_transform(docs_list, embeddings=legal_embeddings)

# Store topic results in the sampled DataFrame
sampled_docs['topic'] = topics

# Print top discovered topics
topic_info = topic_model.get_topic_info()
print(topic_info.head(10))
```

```python
import pandas as pd
import re
import nltk
from collections import Counter
from nltk.corpus import stopwords
from nltk.util import ngrams
from itertools import chain

# Download stopwords if not available
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Function to clean and tokenize text
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation
    words = text.split()  # Tokenize
    words = [word for word in words if word not in stop_words]  # Remove stopwords
    return words

# Apply preprocessing to long_description
docs['cleaned_text'] = docs['long_description'].dropna().apply(preprocess_text)

# Flatten list of all words
all_words = list(chain(*docs['cleaned_text']))

# Count most common words
word_freq = Counter(all_words).most_common(50)  # Top 50 words

# Extract bigrams and trigrams
bigrams = list(ngrams(all_words, 2))
trigrams = list(ngrams(all_words, 3))

# Count most common phrases
bigram_freq = Counter(bigrams).most_common(50)  # Top 50 bigrams
trigram_freq = Counter(trigrams).most_common(50)  # Top 50 trigrams

# Convert to DataFrame for better visualization
df_word_freq = pd.DataFrame(word_freq, columns=['Word', 'Frequency'])
df_bigram_freq = pd.DataFrame([" ".join(bigram) for bigram, _ in bigram_freq], columns=['Bigram'])
df_bigram_freq['Frequency'] = [freq for _, freq in bigram_freq]

df_trigram_freq = pd.DataFrame([" ".join(trigram) for trigram, _ in trigram_freq], columns=['Trigram'])
df_trigram_freq['Frequency'] = [freq for _, freq in trigram_freq]

# Display results
print("\nTop 50 Words:\n", df_word_freq)
print("\nTop 50 Bigrams:\n", df_bigram_freq)
print("\nTop 50 Trigrams:\n", df_trigram_freq)
```
