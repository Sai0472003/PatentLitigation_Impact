```python
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

patents = pd.read_csv('/content/drive/MyDrive/ECO225_Final_Project/PART1/patents.csv')
cases = pd.read_csv('/content/drive/MyDrive/ECO225_Final_Project/PART1/cases.csv')
assignee = pd.read_csv('/content/drive/MyDrive/ECO225_Final_Project/PART1/assignee.csv')
documentid_admin = pd.read_csv('/content/drive/MyDrive/ECO225_Final_Project/PART1/documentid_admin.csv')
```

```python
patents.head()
```

```python
cases.head()
```

```python
cases = cases[['case_row_id', 'case_number', 'court_name', 'date_filed', 'date_closed']]
patents = patents[['case_row_id', 'patent', 'patent_doc_type']]
```

```python
assignee.head()
assignee = assignee[['rf_id', 'ee_name', 'ee_city']]
documentid_admin = documentid_admin[['rf_id', 'grant_doc_num']]
```

```python
assignment_merged = pd.merge(assignee, documentid_admin, on='rf_id', how='inner')
assignment_merged
```

```python
names = pd.read_csv('/content/drive/MyDrive/ECO225_Final_Project/PART1/names.csv')
names =names[['case_row_id', 'name_long', 'party_type']]
names.head()
```

```python
patents_merged = pd.merge(cases, names, on='case_row_id', how='inner')
patents_merged = pd.merge(patents_merged, patents, on='case_row_id', how='inner')
patents_merged
```

```python
assignment_merged['patent'] = assignment_merged['grant_doc_num']
assignment_merged
```

```python
# Filter assignments_merged to include only matching 'patent' values
assignment_merged = assignment_merged[assignment_merged['patent'].isin(patents_merged['patent'].unique())]

# Verify the size of the filtered dataset
print(assignment_merged.shape)
```

```python
complete_merged = pd.merge(assignment_merged, patents_merged, on='patent', how='inner')
complete_merged
```
